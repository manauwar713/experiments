```python
import numpy as np
```

```python
import pandas as pd
data = pd.read_csv('/raid/syscon/malam/experiments/data/raw/adult.csv')
data.head()
data.shape
```

```python
data['workclass']=data['workclass'].replace('?',np.nan)
data['occupation']=data['occupation'].replace('?',np.nan)
data['native-country']=data['native-country'].replace('?',np.nan)
```

```python
df = data.copy()
df.dropna(how='any',inplace=True)
df
```

```python
df.shape
```

```python
df = df.drop_duplicates()
df.shape
```

```python
df = df.drop(['education', 'capital-gain', 'capital-loss'], axis=1)
```

```python
df.shape
```

```python
df.head()
```

```python
df.to_csv('/raid/syscon/malam/experiments/data/raw/cleaned_adult.csv', index=False)
```

```python
import os
for file in os.listdir('/raid/syscon/malam/experiments/data/interim'):
    if file == '.gitkeep':
        continue
    else:
        df = pd.read_csv(os.path.join('/raid/syscon/malam/experiments/data/interim',file))
        gender_counts = df['gender'].value_counts()
        total_samples = gender_counts.sum()
        if len(gender_counts) == 1:
            continue
        else:
            male_proportion = gender_counts[1] / total_samples
            female_proportion = gender_counts[0] / total_samples
            
            
            print(f'Proportion of Male- {male_proportion},proportion of female-{female_proportion}')
    
    
    
```
